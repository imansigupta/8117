```python
# String functions

name = "Mansi"
age = 21
# concatenate
print("My name is "+ name)
print("My name is "+name+"and age is "+str(age))
# formating of string
# 1. Arguments by Position
print("my name is {name} and I am {age}".format(name = name, age = age) )

# 2. f-string (3.6+)
# put F in start
print(f"my name is {name} and I am {age}")

# String Methods

s= 'mansi world'
# s= 'mansiworld'

# Capitalise

print(s.capitalize())

# Uppercase

print(s.upper())

# lower

print(s.lower())


# Swapcase

print(s.swapcase())

# Get Length

print(len(s))

# Replace

print(s.replace('world','everyone'))

# count

sub = 'i'
print(s.count(sub))
# count the no of particular characters in a string 

# starts with return a true or false value

print(s.startswith('mansi'))

# ends with

print(s.endswith('d'))
# returns true or false

# split into a list this split the string into a list and returns it and list is basically an array and it will have all the words in the list
print(s.split())

# find position of the character
print(s.find('r'))

# is all alphanumeric returns true and false
print(s.isalnum())

# is alpha 
print(s.isalpha())


# is all numeric
print(s.isnumeric())

# if we remove space from the words then it becomes alpha numeric and alpha
```

```python
# Creating a list
# number = [1,2,3,4,5]

# # use a constructor

# number2 = list((1,2,3,4,5))
    
# print(number, number2)


# making a list and applying functions on it

fruits = ['apples','oranges','mango','grapes','pears']

# Get a single value from list 
print(fruits)
print(fruits[1])

# get length of the list

print(len(fruits))

# append a fruit in the list

fruits.append('banana')
print(fruits)

# remove from list

fruits.remove('grapes')
print(fruits)

# insert into  a specific position

fruits.insert(2,'cherry')
print(fruits)

# remove with pop
fruits.pop(2)
print(fruits)
# changing the value
fruits[0] = 'blueberries'
print(fruits)

#reverse the list
fruits.reverse()
print(fruits)


# sort the list
# sorts in the alphabateical order
fruits.sort()
print(fruits)

# sort in the reverse order
# sorts in reverse alphabateical order
fruits.sort(reverse=True)
print(fruits)
```

```python
# A tuple is a collection which is ordered and unchangeable. Allows duplicate members

# create a tupple
# creating a tupple require parenthesis
 
fruits = ('Apples','Grapes','Oranges')

# using construcutor
fruits2 = tuple(('Apples','Grapes','Oranges'))

# single value needs trailing comma otherwise it is considered as if it is a normal string or a number value

# fruits3 = ('Apples')
fruits3 = ('Apples',) 
# now considered a tupple
print(fruits, fruits2)
print(fruits3 , type(fruits3))

# getting the value from the tupple

print(fruits[1])

#in tuples can't change value
# fruits[0] = "pears"


# deleting a tupple
del(fruits2)

# print(fruits2)

# Get length of tupple
print(len(fruits))



# A set is a collection which is unordered and unindexed. No duplicate members


# Create ser
print("\n \n Set infromation")
fruits_set = {'apples','oranges','mango'}
print(fruits_set)

# check if in set

print('apples' in fruits_set )

# add to set
fruits_set.add('grapes')
print(fruits_set)

# remove from set
fruits_set.remove('grapes')
print(fruits_set)

# clear set
fruits_set.clear()

# delete set

del fruits_set

# print(fruits_set)
```
